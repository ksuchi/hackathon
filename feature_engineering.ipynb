# after initial EDA, lets train xgboost model with following set of features

```python
# import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import datetime
import seaborn as sns
import numpy as np
from sklearn.metrics import *
import xgboost as xgb
import operator

%matplotlib inline
sns.set()
```

```python
# import datasets
st = datetime.datetime.now()
train_data = pd.read_csv("../data/train.csv")
test_data = pd.read_csv("../data/test_QkPvNLx.csv")
print("total time req to load data ",datetime.datetime.now() - st)
```

```python
input_features = ["Course_Domain","Course_Type","Short_Promotion","Public_Holiday"]
target = ["Sales"]
categorical_features = ["Course_Domain","Course_Type"]
input_cols = ["Course_Domain","Course_Type","Short_Promotion","Public_Holiday","Sales"]
```

## One Hot Categorical Features

### combine training and testing data before one hot encoding

```python
train_data["Flag"] = "train"
test_data["Flag"] = "test"
train_test_data = pd.concat([train_data,
                            test_data])
train_test_data.shape
```

```python
train_data.shape[0]+test_data.shape[0],train_test_data.shape
```

```python
train_test_data.isnull().sum()
```

```python
st = datetime.datetime.now()
one_hot_train_test_data = pd.get_dummies(train_test_data,columns=categorical_features)
print("Time taken for One hot encoding ",datetime.datetime.now() - st)
one_hot_train_test_data.shape
```

```python
one_hot_train_test_data
```

```python
# separate one hot train test data

one_hot_train_data = one_hot_train_test_data[one_hot_train_test_data.Flag=="train"]
one_hot_test_data = one_hot_train_test_data[one_hot_train_test_data.Flag=="test"]

one_hot_train_data.shape,train_data.shape,one_hot_test_data.shape,test_data.shape
```

```python
input_features
```

```python
cols_to_exclude = ['Competition_Metric', 'Course_ID', 'Day_No', 'Flag', 'ID','Long_Promotion', 'Sales','User_Traffic']
```

```python
cols_to_exclude
```

## Model Preparation

```python
# convert train data to the xgboost train object

xgb_train_data = xgb.DMatrix(one_hot_train_data[one_hot_train_data.columns.difference(cols_to_exclude)],one_hot_train_data[target])
xgb_test_data = xgb.DMatrix(one_hot_test_data[one_hot_test_data.columns.difference(cols_to_exclude)])
```

```python
# initial training params
train_params = {"objective" : "reg:linear", "eval_metric" : "rmse"}
```

## train model

```python

xgb_model = xgb.train(params=train_params,dtrain=xgb_train_data)
```

## check feature importance by model

```python
sorted(xgb_model.get_fscore().items(),key=operator.itemgetter(1),reverse=True)
```

```python
np.sqrt(mean_squared_error(one_hot_train_data[target],
                           xgb_model.predict(xgb.DMatrix(one_hot_train_data[one_hot_train_data.columns.difference(cols_to_exclude)])))
       )
```

## lets use feature combinations and retrain model

## experiment 1

### train model by individual feature and check rmse on training data itself

```python
# initial training params
train_params = {"objective" : "reg:linear", "eval_metric" : "rmse"}

for feature in input_features:
    print("\ninput feature for training ",feature)
#     model preparation
    # convert train data to the xgboost train object

    xgb_train_data = xgb.DMatrix(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(feature)],
                                one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])])
    xgb_test_data = xgb.DMatrix(one_hot_test_data.loc[:,one_hot_test_data.columns.str.contains(feature)])
    
    xgb_model = xgb.train(params=train_params,dtrain=xgb_train_data)
    print("feature importance ", sorted(xgb_model.get_fscore().items(),key=operator.itemgetter(1),reverse=True))
    
    print("Training RMSE ",np.sqrt(mean_squared_error(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])],
                           xgb_model.predict(xgb.DMatrix(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(feature)]
                                                        )
                                            )
                                                     )
                                  )
       )
```

## experiment 2

### taken feature combination of 2 

```python
list(itertools.combinations(input_features,2))
```

```python
# initial training params
train_params = {"objective" : "reg:linear", "eval_metric" : "rmse"}

for feature in list(itertools.combinations(input_features,2)):
    print("\ninput feature for training ",feature)
#     model preparation
    # convert train data to the xgboost train object

    xgb_train_data = xgb.DMatrix(one_hot_train_data.loc[:,(one_hot_train_data.columns.str.contains(feature[1]))|(one_hot_train_data.columns.str.contains(feature[0])) ],
                                one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])])
    xgb_test_data = xgb.DMatrix(one_hot_test_data.loc[:,(one_hot_test_data.columns.str.contains(feature[0]))|one_hot_test_data.columns.str.contains(feature[1])])
    
    xgb_model = xgb.train(params=train_params,dtrain=xgb_train_data)
    print("feature importance ", sorted(xgb_model.get_fscore().items(),key=operator.itemgetter(1),reverse=True))
    
    print("** Training RMSE ",np.sqrt(mean_squared_error(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])],
                           xgb_model.predict(xgb.DMatrix(one_hot_train_data.loc[:,(one_hot_train_data.columns.str.contains(feature[1]))|(one_hot_train_data.columns.str.contains(feature[0])) ]
                                                        )
                                            )
                                                     )
                                  )
       )
    
    test_predictions  = xgb_model.predict(xgb_test_data)
    one_hot_test_data["Sales"] = test_predictions
    test_submission = one_hot_test_data[["ID","Sales"]]
    test_submission["Sales"] = round(test_submission["Sales"])
    test_submission.to_csv("../data/test_predictions"+"_with_feature_combinations_"+str(feature)+".csv",
                      header=True,
                      index=False)
```

```python
# initial training params
train_params = {"objective" : "reg:linear", "eval_metric" : "rmse"}

for feature in list(itertools.combinations(input_features,3)):
    print("\ninput feature for training ",feature)
#     model preparation
    # convert train data to the xgboost train object

    xgb_train_data = xgb.DMatrix(one_hot_train_data.loc[:,(one_hot_train_data.columns.str.contains(feature[0]+"|"+feature[1]+"|"+feature[2]))],
                                one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])])
    xgb_test_data = xgb.DMatrix(one_hot_test_data.loc[:,(one_hot_test_data.columns.str.contains(feature[0]+"|"+feature[1]+"|"+feature[2]))])
    
    xgb_model = xgb.train(params=train_params,dtrain=xgb_train_data)
    print("feature importance ", sorted(xgb_model.get_fscore().items(),key=operator.itemgetter(1),reverse=True))
    
    print("** Training RMSE ",np.sqrt(mean_squared_error(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])],
                           xgb_model.predict(xgb.DMatrix(one_hot_train_data.loc[:,(one_hot_train_data.columns.str.contains(feature[0]+"|"+feature[1]+"|"+feature[2]))]
                                                        )
                                            )
                                                     )
                                  )
       )
    test_predictions  = xgb_model.predict(xgb_test_data)
    one_hot_test_data["Sales"] = test_predictions
    test_submission = one_hot_test_data[["ID","Sales"]]
    test_submission["Sales"] = round(test_submission["Sales"])
    test_submission.to_csv("../data/test_predictions"+"_with_feature_combinations_"+str(feature)+".csv",
                      header=True,
                      index=False)
```

## lets do parameter tuning

### vary number of training iterations

### with all features

```python
# initial training params
train_params = {"objective" : "reg:linear", "eval_metric" : "rmse"}
    
# convert train data to the xgboost train object
for iterations in list(range(5,105,5)):
    print("iterations ",iterations)
    xgb_train_data = xgb.DMatrix(one_hot_train_data[one_hot_train_data.columns.difference(cols_to_exclude)],one_hot_train_data[target])
    xgb_test_data = xgb.DMatrix(one_hot_test_data[one_hot_test_data.columns.difference(cols_to_exclude)])
    
    xgb_model = xgb.train(params=train_params,dtrain=xgb_train_data,num_boost_round=iterations)
    print("feature importance ",sorted(xgb_model.get_fscore().items(),key=operator.itemgetter(1),reverse=True))
    print("*** Training RMSE ",np.sqrt(mean_squared_error(one_hot_train_data[target],
                               xgb_model.predict(xgb.DMatrix(one_hot_train_data[one_hot_train_data.columns.difference(cols_to_exclude)])))
           ))
    test_predictions  = xgb_model.predict(xgb_test_data)
    one_hot_test_data["Sales"] = test_predictions
    test_submission = one_hot_test_data[["ID","Sales"]]
    test_submission["Sales"] = round(test_submission["Sales"])
    test_submission.to_csv("../data/test_predictions"+"_with_boosting_rounds_"+str(iterations)+".csv",
                      header=True,
                      index=False)
```

## with individual features

```python
# initial training params
train_params = {"objective" : "reg:linear", "eval_metric" : "rmse"}

for iterations in list(range(5,105,5)):
    print("--- iterations ",iterations)
    for feature in input_features:
        print("\n---- input feature for training ",feature)
    #     model preparation
        # convert train data to the xgboost train object

        xgb_train_data = xgb.DMatrix(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(feature)],
                                    one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])])
        xgb_test_data = xgb.DMatrix(one_hot_test_data.loc[:,one_hot_test_data.columns.str.contains(feature)])

        xgb_model = xgb.train(params=train_params,dtrain=xgb_train_data,num_boost_round=iterations)
        print("feature importance ", sorted(xgb_model.get_fscore().items(),key=operator.itemgetter(1),reverse=True))

        print("****  Training RMSE ",np.sqrt(mean_squared_error(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(target[0])],
                               xgb_model.predict(xgb.DMatrix(one_hot_train_data.loc[:,one_hot_train_data.columns.str.contains(feature)]
                                                            )
                                                )
                                                         )
                                      )
           )
        test_predictions  = xgb_model.predict(xgb_test_data)
        one_hot_test_data["Sales"] = test_predictions
        test_submission = one_hot_test_data[["ID","Sales"]]
        test_submission["Sales"] = round(test_submission["Sales"])
        test_submission.to_csv("../data/test_predictions"+"_with_boosting_rounds_"+str(iterations)+"_with_feature_"+str(feature)+".csv",
                          header=True,
                          index=False)
```
